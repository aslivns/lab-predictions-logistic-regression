### X-y split

```python
# Build X and y
y = data_all['target_col']
X = data_all.drop(['target_col'], axis=1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
# from sklearn.preprocessing import StandardScaler

X_train_num = X_train.select_dtypes(include = np.number)

# Scaling data
transformer = MinMaxScaler().fit(X_train_num) # need to keep transformer
X_train_normalized = transformer.transform(X_train_num)
X_train_norm = pd.DataFrame(X_train_normalized)
```

```python
X_train.columns = X_train.columns
X_train.head(10)
```

```python
X_train_categorical = X_train.select_dtypes(include = np.object)
X_train_categorical.head()
```

```python
X_train_norm.columns = X_train_num.columns
X_train_norm.head()
```

```python
X_train_categorical.columns
```

```python
X_train_categorical = X_train.select_dtypes(include = np.object)
X_train_cat = pd.get_dummies(X_train_categorical, 
                             columns=['title', 'rating', 'special_features', 'category', 'language'],
                             drop_first=True)
X_train_cat.head()
```

```python
X_train_transformed = np.concatenate([X_train_norm, X_train_cat], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
classification = LogisticRegression(random_state=1234, solver='lbfgs')
classification.fit(X_train_transformed, y_train)
# classification = LogisticRegression(random_state=0, solver='saga',
#                   multi_class='multinomial').fit(X_train_transformed, y_train)
```

```python
# for numericals
X_test_num = X_test.select_dtypes(include = np.number)

# Scaling data
# we use the transformer that was trained on the training data
X_test_normalized = transformer.transform(X_test_num)
X_test_norm = pd.DataFrame(X_test_normalized)
```

```python
# for categoricals
X_test_categorical = X_test.select_dtypes(include = np.object)
X_test_cat = pd.get_dummies(X_test_categorical, 
                            columns=['title', 'rating', 'special_features', 'category', 'language'],
                            drop_first=True)
X_test_cat.head()
```

```python
X_test_transformed = np.concatenate([X_test_norm, X_test_cat], axis=1)
```

```python
# Now we can make predictions on the test set:
predictions = classification.predict(X_test_transformed)
classification.score(X_test_transformed, y_test)
```
